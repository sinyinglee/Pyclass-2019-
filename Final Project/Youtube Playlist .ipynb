## Step 1-1: 建立帳號認證的function

```python
# -*- coding: utf-8 -*-

# Sample Python code for youtube.search.list
# See instructions for running these code samples locally:
# https://developers.google.com/explorer-help/guides/code_samples#python

import os

import google_auth_oauthlib.flow
import googleapiclient.discovery
import googleapiclient.errors

scopes = ["https://www.googleapis.com/auth/youtube.force-ssl"]
os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"
api_service_name = "youtube"
api_version = "v3"
client_secrets_file = "client.json"

# Get credentials and create an API client
def authorize():
    flow = google_auth_oauthlib.flow.InstalledAppFlow.from_client_secrets_file(
        client_secrets_file, scopes)
    credentials = flow.run_console()
    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, credentials=credentials)
    return youtube
```

## Step 2-1 建立各個function

```python
#搜尋歌曲
def request(query, youtube):
    request = youtube.search().list(
    part=" snippet",
    q=query)
    response = request.execute()
    return response

#建立播放清單
def request_playlist(p_title):
    request = youtube1.playlists().insert(
        part="snippet",
        body={
            "snippet": {
            "title":p_title}
        })
    response = request.execute()
    return response


# 將歌曲輸入至播放清單
def insert_item(v_id):
    request = youtube.playlistItems().insert(
        part="snippet",
        body={
          "snippet": {
            "playlistId": playlist_id,
            "position": 0,
            "resourceId": {
              "kind": "youtube#video",
              "videoId": v_id}}}
    )
    response = request.execute()
    playlist_url = f"https://www.youtube.com/playlist?list={playlist_id}"
    return playlist_url
```

## Step 2-2 執行程式

```python
song_name = input("請輸入歌曲名稱： ")
playlist_title = input("請輸入播放清單名稱： ")

playlist_id = request_playlist(playlist_title)["id"]

song_list = song_name.split(',')
for i in song_list:
    # 使用youtube API 搜尋歌曲
    get_song = request(i, youtube1)
    v_id = get_song["items"][0]["id"]["videoId"]
    play_url = insert_item(v_id)
    
print(play_url)

#以下出現error是因為已超過今日使用量
#正常會跳出一個一個url即為你的playlist連結
```
