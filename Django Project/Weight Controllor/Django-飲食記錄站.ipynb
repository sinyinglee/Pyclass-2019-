### 使用飲食記錄站

1. 於終端機cd Weight Controllor
2. 下載虛擬環境,於終端機下指令 
    --> pip install virtualenv
3. cd my_app
4. 建立與啟動虛擬環境
    --> virtualenv test_env01(虛擬環境檔案名稱）
    --> source test_env01/bin/activate （啟動虛擬環境）
5. 下載Django
   --> pip install Django
6. 啟動網頁
   --> python manage.py runserver
   (複製於終端機顯示的連結，開啟網頁）
