### 使用map

```python
items = [500, 30, 200]
c = 30.8594
```

```python
egg = []
for item in items:
    p = item*c
    egg.append(p)
```

```python
egg
```

```python
def exchange(i):
    return i*c
```

```python
exchange(100)
```

```python
new = list(map(exchange, items))
```

```python
new
```

### 使用lambda(救急用函數)

```python
list(map(lambda p:p*c, items))
```

### list comprehension

```python
[p*c for p in items]
```

## 數字照我們的意思顯示

```python
c = 30.859435 #1 dolllar to 1 NTD
```

```python
print(f"1美金合台幣{c}元")
```

```python
print(f"1美金合台幣{c:.2f}元") #小數點後二位,浮點數
```

```python
print(f"1美金合台幣{c:10.2f}元") #數字前面印出10個空格, 取小數點後兩位
```

```python
print(f"1美金合台幣{c:*>10.3f}元") #數字靠右邊印出10個＊, 取小數點後三位
```
