## 詞評統計機器人
### 使用手冊
1. 輸入歌詞
2. 輸入===, 結束程式


```python
song_list=[]
while (True):
    line=input()
    line_list=line.split()
    if line =="===":
        break
    for data in line_list:
        data=data.lower()
        data=data.replace(",", "")
        data=data.replace(".", "")
        data=data.replace(":", "")
        data=data.replace("?", "")
        data=data.replace("\n", "")
        
        if data=="We're":
            song_list.append("we")
            song_list.append("are")
        elif data=="I'd":
            song_list.append("I")
            song_list.append("would")
        else:
            song_list.append(data)

song_list.sort()
            
count_lyric=[]
for data in song_list:
    if data in count_lyric:
        continue
    else:
        count_lyric.append(data)
        print("{}:{}".format(data, song_list.count(data)))
                
```
