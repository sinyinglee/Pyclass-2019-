```python
import requests
products = []
for i in range(1,6):
    res = requests.get(f'https://search.taobao.tw/catalog/?_keyori=ss&ajax=true&from=input&page={i}&q=iphone')
    jd = res.json()
    df = pandas.DataFrame(jd['mods']['listItems'])
    products.append(df)
```

```python
df_all = pandas.concat(products)
```
