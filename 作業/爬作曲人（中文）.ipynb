```python
def find_lyrics():
    table=[["歌名","作詞","作曲"]]
    import requests
    from pyquery import PyQuery as pq
    for i in chinese_outer_list:
            song_name = i[0]
            print_list=[]
            print_list.append(song_name)
            try:
                response = requests.get("https://mojim.com/{}.html?t3".format(song_name)) 
                doc= pq(response.text)
                doc.make_links_absolute(base_url = response.url)   
                get_url = doc(" span.mxsh_ss4 > a").attr("href")
                response = requests.get(get_url) 
                doc= pq(response.text)
                get_lyric = doc("#fsZx3").text()
                make_list = get_lyric.split("\n")
                lyric=make_list[0].replace("作詞：", "")
                melody=make_list[1].replace("作曲：", "")
                print_list.append(lyric)
                print_list.append(melody)
            except:
                print_list.append(" ")
                print_list.append(" ")
            table.append(print_list)
    return table
            
print_table= find_lyrics()

```

```python
import csv 
with open("chinese_finish_doc.csv", "w") as fOut:
    csvOut=csv.writer(fOut)
    csvOut.writerows(print_table)
```
